## Exemplo Resolvido 1: Ano Bissexto

**O problema:** Crie um programa que peça um ano (ex: 2024) e informe se ele é Bissexto ou não.

*Regra do Ano Bissexto:*
Ele precisa ser múltiplo de 4 (`ano % 4 == 0`).
Mas se for fim de século (múltiplo de 100), ele **não** é bissexto, a menos que ele também seja múltiplo de 400.

```python
# Solução do Professor
ano = int(input("Digite o ano: "))

if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} É bissexto!")
else:
    print(f"O ano {ano} NÃO é bissexto.")
```

## Exemplo Resolvido 2: Calculadora Simples

**O problema:** Peça dois números e um operador matemático (`+`, `-`, `*` ou `/`) ao usuário. Imprima o resultado da operação.

```python
num1 = float(input("Digite o primeiro número: "))
op = input("Digite a operação (+, -, *, /): ")
num2 = float(input("Digite o segundo número: "))

if op == "+":
    print(f"Resultado: {num1 + num2}")
elif op == "-":
    print(f"Resultado: {num1 - num2}")
elif op == "*":
    print(f"Resultado: {num1 * num2}")
elif op == "/":
    # Cuidados extras em programação: não se divide por zero!
    if num2 == 0:
        print("Erro: Divisão por zero.")
    else:
        print(f"Resultado: {num1 / num2}")
else:
    print("Operador inválido!")
```

- Equilátero: 3 lados iguais;
- Isósceles: 2 lados iguais;
- Escaleno: 3 lados diferentes.
*Dica: Para formar um triângulo, a soma de quaisquer dois lados precisa ser maior que o terceiro lado.*

```python
# Seu código aqui
print ("Por favor digite os três lados do triangulo.")
lado1 = float(input("Digite o primeiro lado: "))
lado2 = float(input("Digite o segundo lado: "))
lado3 = float(input("Digite o terceiro lado: "))

if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
  print ("Os valores formam um triangulo.")

  if lado1 == lado2 == lado3:
    print ("O triangulo é Equilatero.")

  elif lado1 == lado2 or lado1 == lado3 or lado2 == lado1 or lado2 == lado3 or lado3 == lado1 or lado3 == lado2:
    print ("O triangulo é Isósceles.")

  else:
    print ("O triangulo é Escaleno.")

else:
  print ("Os valores não formam um triangulo")
```

### Exercício 9: Sistema de Climatização
Um ar-condicionado liga se:
- A temperatura for >= 25°C **E** houver pessoas na sala.
- **OU** se a temperatura for >= 30°C (mesmo sem pessoas).
- **NÃO** liga se a janela estiver aberta.

```python
print("Sistema automático de climatização")

temperatura = float(input("Qual é a temperatura da sala?: "))
pessoas = input("Tem pessoas na sala? sim ou não: ").lower()
janela = input("A janela está aberta? sim ou não: ").lower()

if (janela == "sim"):
    print("Ar-condicionado não liga.")

elif (temperatura >= 30 and janela == "não"):
    print("Ar-condicionado liga.")

elif (temperatura >= 25 and pessoas == "sim" and janela == "não"):
    print("Ar-condicionado liga.")

else:
    print("Ar-condicionado continua desligado.")
```
